# Ind√≠ce de pessoas descoupadas/ocupadas antes e depois do Amigos do Bem atuarem

Foram an√°lisada as seguintes informa√ß√µes do Censo dos anos de 2000 e 2010:

- Popula√ß√£o economicamente ativa
- Taxa de desocupa√ß√£o
- Taxa de atividade

```python
import pandas as pd

file_path = "/content/drive/MyDrive/BCG X Challenge/Dados/Censo/Censo_municipal_estadual_nacional.xlsx"
sheet_name = "MUN 91-00-10"
municipios_amigos_do_bem = ["MAURITI", "INAJ√Å", "BU√çQUE", "S√ÉO JOS√â DA TAPERA"]

df = pd.read_excel(file_path, sheet_name=sheet_name)
filtro = df['Munic√≠pio'].isin(municipios_amigos_do_bem)
df_municipios_amigos_do_bem = df[filtro]
pd.set_option('display.max_rows', None)
df_municipios_amigos_do_bem
```

### Popula√ß√£o Economicamente Ativa

Popula√ß√£o economicamente ativa (PEA) corresponde ao n√∫mero de pessoas nessa faixa et√°ria que, na semana de refer√™ncia do Censo, encontravam-se ocupadas no mercado de trabalho ou que, encontrando-se desocupadas, tinham procurado trabalho no m√™s anterior √† data da pesquisa.

```python
import matplotlib.pyplot as plt

def pea_plot(column, start, end="ou mais", position=None):
  plt.figure(figsize=(10,6))
  end = f"a {end} anos" if end.isdigit() else end
  for municipio in municipios_amigos_do_bem:
      subset = df_municipios_amigos_do_bem[df_municipios_amigos_do_bem['Munic√≠pio'] == municipio]
      plt.plot(subset['ANO'], subset[column], label=municipio, marker='o')

  plt.title(f"Popula√ß√£o Economicamente Ativa (PEA) com {start} anos {end} de acordo com Censo 2000 e 2010")
  plt.xlabel("Ano")
  plt.ylabel("PEA")
  plt.xticks([2000, 2010])
  plt.legend()
  plt.grid(True)
  plt.tight_layout()
  plt.show()

plt.figure(figsize=(15,10))

pea_plot('PEA', '10', 'ou mais', position=1)
pea_plot('PEA1014', '10', '14', position=3)
pea_plot('PEA1517', '15', '17', position=4)
pea_plot('PEA18M', '18', position=2)
```

### Taxa de desocupa√ß√£o

Taxa de desocupa√ß√£o se refere ao porcentual da popula√ß√£o economicamente ativa (PEA) na faixa et√°ria estudada que estava desocupada,  ou seja, que n√£o estava ocupada na semana anterior √† data do Censo mas havia procurado trabalho no longo do m√™s anterior √† data dessa pesquisa

```python
import matplotlib.pyplot as plt

def unemployment_plot(column, start, end="ou mais", label="Desocupa√ß√£o", position=None):
  plt.figure(figsize=(10,6))
  end = f"a {end} anos" if end.isdigit() else end
  for municipio in municipios_amigos_do_bem:
      subset = df_municipios_amigos_do_bem[df_municipios_amigos_do_bem['Munic√≠pio'] == municipio]
      plt.plot(subset['ANO'], subset[column], label=municipio, marker='o')

  plt.title(f"Taxa de {label} de Pessoas com {start} anos {end} de acordo com Censo 2000 e 2010")
  plt.xlabel("Ano")
  plt.ylabel(f"Taxa de {label} (%)")
  plt.xticks([2000, 2010])
  plt.legend()
  plt.grid(True)
  plt.tight_layout()
  plt.show()

plt.figure(figsize=(15,10))

unemployment_plot('T_DES', '10', 'ou mais', position=1)
unemployment_plot('T_DES18M', '18', position=2)
unemployment_plot('T_DES1014', '10', '14', position=3)
unemployment_plot('T_DES1517', '15', '17', position=4)
unemployment_plot('T_DES1824', '18', '24', position=5)
unemployment_plot('T_DES2529', '25', '29', position=6)
```

### Taxa de atividade

Raz√£o entre as pessoas que eram economicamente ativas, ou seja, que estavam ocupadas ou desocupadas na semana de refer√™ncia do Censo e o total de pessoas nesta faixa et√°ria multiplicado por 100. Considera-se desocupada a pessoa que, n√£o estando ocupada na semana de refer√™ncia, havia procurado trabalho no m√™s anterior a essa pesquisa.

```python
unemployment_plot('T_ATIV', '10', 'ou mais', label='Atividade', position=1)
unemployment_plot('T_ATIV18M', '18', label='Atividade', position=2)
unemployment_plot('T_ATIV1014', '10', '14', label='Atividade', position=3)
unemployment_plot('T_ATIV1517', '15', '17', label='Atividade', position=4)
unemployment_plot('T_ATIV1824', '18', '24', label='Atividade', position=5)
unemployment_plot('T_ATIV2529', '25', '29', label='Atividade', position=6)
```
